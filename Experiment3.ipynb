```python
'''
描述并实现动态规划的作业排程算法
'''
s1 = [2,7,9,3,4,80,3]
s2 = [4,8,5,6,4,5,6]
t1 = [999,2,3,1,3,999]
t2 = [999,2,1,2,2,999]

final_track = []
l1 = [0,0,0,0,0,0,0]
l2 = [0,0,0,0,0,0,0]

def assignmentQuestion_s1(j):
    if j <= 0:
        return 0
    if j == 1:
        return s1[0] + s1[1]
    else:
        result = min(assignmentQuestion_s1(j-1)+s1[j],  assignmentQuestion_s2(j-1)+t2[j-1]+s1[j])
#         final_track.append(result)
        return result
    
def assignmentQuestion_s2(j):
    if j <= 0:
        return 0
    if j == 1:
        return s2[0] + s2[1]
    else:
        result = min(assignmentQuestion_s2(j-1)+s2[j],  assignmentQuestion_s1(j-1)+t1[j-1]+s2[j])
#         final_track.append(result)
        return result
    
def shortestPath(n):
    if n<=0 or n>6:
        return -1
    return min(assignmentQuestion_s1(n),assignmentQuestion_s2(n))


for i in range(1,7):
    
    result1 = min(assignmentQuestion_s1(i-1)+s1[i],  assignmentQuestion_s2(i-1)+t2[i-1]+s1[i])
    result2 = min(assignmentQuestion_s2(i-1)+s2[i],  assignmentQuestion_s1(i-1)+t1[i-1]+s2[i])
    
    print(i,":",result1,result2)
    if result1 < result2:
        final_track.append("s1"+str(i))
    else:
        final_track.append("s2"+str(i))

# print(final_track) # wrong



for i in range(1,7):
    if assignmentQuestion_s1(i) < assignmentQuestion_s2(i-1)+t2[i-1]+s1[i]:
        l1[i] = 1
    else:
        l1[i] = 2
        
    if assignmentQuestion_s2(i-1)+s2[i] < assignmentQuestion_s1(i-1)+t1[i-1]+s2[i]:
        l2[i] = 2
    else:
        l2[i] = 1
        
        
        
print(l1,l2)
# [0, 1, 1, 2, 1, 1, 1] 
# [0, 2, 1, 2, 1, 2, 2]
flag = 2 
final_track2 = []
for i in range(6,0,-1):
    if flag == 1:
        final_track2.insert(0,"s1"+str(i))
        if l1[i] == 2:
            flag = 2
    else:
        final_track2.insert(0,"s2"+str(i))
        if l2[i] == 1:
            flag = 1
        
print(final_track2)
```

```python
'''
描 述 并 实 现 最 长 共 同 子 序 列 动 态 规 划 算 法 , 
并 显 示 S1=ACCGGTCGAGATGCAG,S2 = GTCGTTCGGAATGCAT 的最长共同子序列。
'''

S1 = "ACCGGTCGAGATGCAG"
S2 = "ACCGGTCGAGATGCAG"
lcs = ""

def LCS(i,j):
    global S1,S2
    
    if i == 0 and j == 0 and S1[0] == S2[0]:
        return 1
    
    if i*j < 0 or i<0 or j<0:
        return -1
    
    if i==0 or j==0:
        return 0
    
    if S1[i]==S2[j] and (i*j)>0:
        return LCS(i-1,j-1)+1
    
    if S1[i]!=S2[j] and (i*j)>0:
        return max(LCS(i,j-1), LCS(i-1,j))
    
for i in range(16):
    print("i = ",i,",LCS = ",LCS(i,i))
    if i==0 and S1[0] == S2[0]:
        lcs+=S1[i]
    if i>0 and LCS(i,i)>LCS(i-1,i-1):
        lcs+=S1[i]
    
print(lcs," length is ",str(len(lcs)))
    
```
